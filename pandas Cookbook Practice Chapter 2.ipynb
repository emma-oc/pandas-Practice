##Selecting column and rows
We can try:

    * Index the names of columns

```python
complaints['Complaint Type']
```

    * Slice the dataframe, get a few rows

```python
complaints[:5]
```

    * or a combination of the above (order of columns and row index doesn't matter)

```python
complaints['Complaint Type'][:5]
```

```python
complaints[:5]['Complaint Type']
```

##Selecting multiple columns
Simply index with list of columns you want.

```python
complaints[['Complaint Type', 'Borough']]
```

```python
complaints[['Complaint Type', 'Borough']][:10]
```

##Most common complaint type?
Use `.value_counts()` method

```python
complaints['Complaint Type'].value_counts()
```

Find out the top 10

```python
complaints['Complaint Type'].value_counts()[:10]
```

And plot a bar graph.

```python
complaints['Complaint Type'].value_counts()[:10].plot(kind='bar')
```
