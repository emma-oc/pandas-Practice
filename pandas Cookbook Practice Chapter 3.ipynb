##Selcting "noisy" complaint : binary array selection

"Noisy" complaints

```python
complaints = pd.read_csv('data/311-service-requests.csv')
```

```python
complaints[:5]
```

Subseting all complaints that are about noise using **boolean indexing**: row filtering by a boolean array
    * length of array and dataframe's index should be the same

```python
noise_complaints = complaints[complaints['Complaint Type'] == "Noise - Street/Sidewalk"]
noise_complaints[:3]
```

Using operation to apply multiple filters:

```python
is_noise = complaints['Complaint Type'] == "Noise - Street/Sidewalk"
in_brooklyn = complaints['Borough'] == "BROOKLYN"
complaints[is_noise & in_brooklyn][:5]
```

If using logic operation in indexing **parentheses are needed**, such as `(df['a']=='filter1') & (df['b']=='filter2')`.

NumPy arrays (of length greater than 1) and Pandas objects such as Series do not have a boolean value -- they will raise error.

```python
complaints[(complaints['Complaint Type'] == "Noise - Street/Sidewalk") & (complaints['Borough'] == "BROOKLYN")]
```

```python
complaints[is_noise & in_brooklyn][['Complaint Type', 'Borough', 'Created Date', 'Descriptor']][:10]
# columns index is a list
```

##Notes on object types : pd.Series and numpy array

Each columns is a `pd.Series`, whici is internally numpy arrays.

Add `.value` at the end of a `pd.Series`, we'll get the internal numpy array.

```python
pd.Series([1,2,3]).values
```

```python
np.array([1,2,3])
```

So binary-array-selection works for any numpy array.

```python
arr = np.array([1,2,3])
arr != 2
```

```python
arr[arr != 2]
```

##Which borough has the most noise complaints?
Binary array selection + `.value_counts()` from previous Chapter

```python
noise_complaints = complaints[complaints['Complaint Type'] == "Noise - Street/Sidewalk"]
noise_complaints['Borough'].value_counts()
```

```python
noise_complaints['Borough'].value_counts().plot(kind='bar')
```

Calculate the ratio:

```python
noise_complaint_counts = noise_complaints['Borough'].value_counts()
complaint_counts = complaints['Borough'].value_counts()
noise_complaint_counts / complaint_counts
# add .astype(float) is need to convert integer to float for division...
```

```python
(noise_complaint_counts / complaint_counts).plot(kind='bar')
```
